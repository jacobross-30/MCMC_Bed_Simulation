```python
from pyproj import Proj, Transformer

# Antarctic polar stereographic projection (EPSG:3031)
proj_ps = Proj('EPSG:3031')
proj_latlon = Proj('EPSG:4326')

# Transformer (EPSG:3031 -> WGS84 lat/lon)
transformer = Transformer.from_proj(proj_ps, proj_latlon, always_xy=True)

# Convert your meshgrid coordinates
lon, lat = transformer.transform(xx, yy)
```

```python
from pyproj import Transformer
proj_ps = Proj('EPSG:3031')      # Antarctic stereographic
proj_ll = Proj('EPSG:4326')      # WGS84 lat/lon
transformer = Transformer.from_proj(proj_ps, proj_ll, always_xy=True)

# Plot in projected x/y (meters)
plt.pcolormesh(xx, yy, vel_mag, shading='auto')
plt.colorbar(label='Velocity (m/yr)')
plt.axis('scaled')
#plt.title('Velocity magnitude (m/yr)')

# Now relabel axes: pick tick locations in meters, then convert to lat/lon
xticks = np.linspace(xx.min(), xx.max(), 6)   # 6 ticks across x
yticks = np.linspace(yy.min(), yy.max(), 6)   # 6 ticks across y

# Convert tick locations to lat/lon
xticks_lon, xticks_lat = transformer.transform(xticks, np.full_like(xticks, yy.mean()))
yticks_lon, yticks_lat = transformer.transform(np.full_like(yticks, xx.mean()), yticks)

# Apply new ticks (in degrees)
plt.xticks(xticks, labels=[f"{lon:.1f}Â°" for lon in xticks_lon])
plt.yticks(yticks, labels=[f"{lat:.1f}Â°" for lat in yticks_lat])

#plt.xlabel("Longitude (Â°)")
#plt.ylabel("Latitude (Â°)")
def add_scalebar(ax, length_km=20, location=(0.85, 0.95), linewidth=3):
    """
    Add scale bar to map.
    length_km : length of scale bar in kilometers
    location  : (x_frac, y_frac) location in axis fraction coords
    """
    # Convert km to meters
    length_m = length_km * 1000

    # Get axis limits in data coords
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()

    # Scale bar start and end in data coords
    x_start = xlim[0] + location[0] * (xlim[1] - xlim[0])
    y_start = ylim[0] + location[1] * (ylim[1] - ylim[0])
    x_end = x_start + length_m

    # Draw line
    ax.plot([x_start, x_end], [y_start, y_start], color="white", lw=linewidth)
    ax.text((x_start + x_end) / 2, y_start - 0.02 * (ylim[1] - ylim[0]),
            f"{length_km} km", ha="center", va="top", fontsize=9, color="white")

# Add scale bar to current axes
ax = plt.gca()
add_scalebar(ax, length_km=20)

plt.savefig("velocity_mag_with_scalebar.png", dpi=300, bbox_inches="tight")
plt.show()
```

The *highvel_mask* returned try to smoothly enclose a region of high velocity and excluding locations where the ice is not grounded. This region could be used later to constrain the sampling location of the MCMC.

The function works as first find high velocity region, then smooth the boundary of this region. Because this smoothing generally will shrink the boundary, the boundary is later expanded outward for *distance_max* meters.

The degree of smoothness of the boundary could be modified by the optional argument *smooth_mode*, which is default to 10. A higher smooth_mode will give a smoother boundary

The *ocean_mask* will be 1 when the location is ocean (open ocean water without ice, sea ice, or ice shelf), and will be 0 otherwise.

The *grounded_ice_mask* is 1 at location where ice is present and grounded, and will be 0 otherwise.

An example where *ocean_mask == 0 and grounded_ice_mask == 0* will be ice-free terrestrial land

__Q3: Change distance_max and smooth_mode parameters, how do you see the high velocity region changed by these parameter? Which parameter give a reasonable answer?__

Answer: when i decreased the max distance and smooth mode, the high velocity region in the chart became more rigid and showed more detail around the edges. The overall shape looks very similar, but there are some purple areas around the edges that aren't there when the distance max is at 3500 and the smooth mode is at 18 

Notice, we aims to cover region with 50 m/yr velocity magnitude

```python
# calculate high velocity region
ocean_mask = (bedmap_mask == 3) | (np.isnan(bedmap_mask)) # utilize the mask in BedMachine dataset to characterize ice regions
grounded_ice_mask = (bedmap_mask == 1) | (bedmap_mask == 2)
distance_max = 3500
velocity_threshold = 50
smooth_mode = 18
highvel_mask = Topography.get_highvel_boundary(velx, vely, velocity_threshold, grounded_ice_mask, ocean_mask, 
                                               distance_max, xx, yy, smooth_mode = smooth_mode)
```

```python
plt.pcolormesh(xx,yy,highvel_mask)
plt.axis('scaled')
#plt.pcolormesh(xx,yy,bedmap_mask, alpha=0.5)
#plt.axis('scaled')
```
